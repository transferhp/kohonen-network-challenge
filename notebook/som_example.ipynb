```python
from kohonen_challenge import SOM
```

```python
def generate_sample_data(rows: int, cols: int, seed: Optional[int]=1234) -> np.ndarray:
    np.random.seed(1234)
    input_data = np.random.random((rows, cols))
    return input_data
```

```python
def render_ouput_map(image_data) -> None:
    plt.imshow(image_data)
```

```python
# Generate random 20 colors data
input_data = generate_sample_data(20, 3)  
```

# Train a 10x10 network over 100 iterations

```python
som = SOM(10,10, 100)
```

```python
%time som.fit(input_data)
```

Training finishes less than 1s, not bad.

```python
# See the output map
render_ouput_map(som.get_map)
```
